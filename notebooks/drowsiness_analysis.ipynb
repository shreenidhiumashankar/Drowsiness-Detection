# Drowsiness & Yawn Metric Threshold Analysis

This interactive notebook demonstrates Eye Aspect Ratio (EAR) and Mouth Aspect Ratio (MAR) metric computation and threshold evaluation using MediaPipe Face Mesh.

```python
from scipy.spatial import distance

def calculate_ear(eye):
    a = distance.euclidean(eye[1], eye[5])
    b = distance.euclidean(eye[2], eye[4])
    c = distance.euclidean(eye[0], eye[3])
    return (a + b) / (2.0 * c)

# Mock landmarks for an open eye
open_eye = [(0, 10), (3, 16), (7, 16), (10, 10), (7, 4), (3, 4)]
print(f"Open Eye EAR: {calculate_ear(open_eye):.2f}")
```
